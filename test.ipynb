```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create sample GDP data
gdp_data = {
    'Country': ['USA', 'China', 'Japan', 'Germany', 'India'],
    'GDP_2020': [20.94, 14.72, 5.06, 3.85, 2.66],
    'GDP_Growth': [3.5, 6.1, 2.2, 1.1, 7.0]
}
df = pd.DataFrame(gdp_data)

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 12), dpi = 100)

# Bar plot of GDP values
axes[0,0].bar(df['Country'], df['GDP_2020'], color='skyblue')
axes[0,0].set_title('GDP by Country (2020)')
axes[0,0].set_ylabel('GDP (Trillion USD)')
axes[0,0].tick_params(axis='x', rotation=45)

# Horizontal bar plot of growth rates
axes[0,1].barh(df['Country'], df['GDP_Growth'], color='lightgreen')
axes[0,1].set_title('GDP Growth Rate (%)')
axes[0,1].set_xlabel('Growth Rate (%)')

# Scatter plot
axes[1,0].scatter(df['GDP_2020'], df['GDP_Growth'], s=100)
for i, txt in enumerate(df['Country']):
    axes[1,0].annotate(txt, (df['GDP_2020'].iloc[i], df['GDP_Growth'].iloc[i]))
axes[1,0].set_xlabel('GDP (Trillion USD)')
axes[1,0].set_ylabel('Growth Rate (%)')
axes[1,0].set_title('GDP vs Growth Rate')

# Pie chart of GDP distribution
axes[1,1].pie(df['GDP_2020'], labels=df['Country'], autopct='%1.1f%%')
axes[1,1].set_title('GDP Distribution')

plt.tight_layout()
plt.show()

# Additional analysis with linear regression
plt.figure(figsize=(10, 6))
z = np.polyfit(df['GDP_2020'], df['GDP_Growth'], 1)
p = np.poly1d(z)
plt.scatter(df['GDP_2020'], df['GDP_Growth'])
plt.plot(df['GDP_2020'], p(df['GDP_2020']), "r--", alpha=0.8)
plt.title('GDP vs Growth Rate (with regression line)')
plt.xlabel('GDP (Trillion USD)')
plt.ylabel('Growth Rate (%)')
plt.show()
```

```python
# Let's create some sample data to practice with
import numpy as np

# Generate random data for monthly sales over 2 years
months = 24
time = np.arange(months)
base_sales = 1000
seasonal_pattern = 200 * np.sin(2 * np.pi * time / 12)  # Seasonal variation
trend = 50 * time  # Upward trend
noise = np.random.normal(0, 100, months)  # Random noise
sales = base_sales + seasonal_pattern + trend + noise

# Generate customer satisfaction scores
satisfaction_scores = 7 + np.random.normal(0, 0.5, months)

# Create market share percentages
market_share = 30 + 0.5 * time + np.random.normal(0, 2, months)
market_share = np.clip(market_share, 0, 100)  # Ensure values stay between 0-100

# Generate competitor data
competitor_sales = base_sales * 0.8 + np.random.normal(0, 150, months)

# Now you have interesting data to work with! Try to:
# 1. Create a line plot showing sales over time with seasonal patterns
# 2. Make a scatter plot comparing sales vs satisfaction scores
# 3. Create a subplot with both sales and competitor sales
# 4. Add markers, colors, and styling to make the visualizations informative
# 5. Try adding legends, grid lines, and proper labels
```

```python
# Create a DataFrame from all the arrays
df = pd.DataFrame({
    'Time': time,
    'Sales': sales,
    'Satisfaction': satisfaction_scores,
    'Market_Share': market_share,
    'Competitor_Sales': competitor_sales
})

# Optional: Add a proper date index starting from January 2022
df['Date'] = pd.date_range(start='2022-01-01', periods=months, freq='M')
df.set_index('Date', inplace=True)

# Display the first few rows
print(df.head())
```

```python
# Create a figure with dimensions 12 inches wide by 6 inches tall
plt.figure(figsize=(12, 6))
# Plot the sales data over time with blue line of width 2 and label for legend
plt.plot(df.index, df['Sales'], color='blue', linewidth=2, label='Monthly Sales')
# Add a title to the plot with font size 12
plt.title('Monthly Sales Over Time (2022-2023)', fontsize=12)
# Label the x-axis with font size 10
plt.xlabel('Date', fontsize=10)
# Label the y-axis with font size 10
plt.ylabel('Sales', fontsize=10)
# Add a grid with dashed lines at 70% opacity
plt.grid(True, linestyle='--', alpha=0.7)
# Display the legend using the label from plot()
plt.legend()
# Rotate x-axis labels 45 degrees for better readability
plt.xticks(rotation=45)
# Adjust the layout to prevent label cutoff
plt.tight_layout()
# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the GDP data
gpd_df = pd.read_csv("gdp__data-Sat Dec 07 2024.csv")
```

```python
gpd_df.head(4)
```

```python
# Filter for Southeast Asian countries
sea_countries = ['Thailand', 'Vietnam', 'Indonesia', 'Malaysia', 'Philippines', 
                 'Singapore', 'Cambodia', 'Myanmar', 'Laos', 'Brunei']
sea_gdp = gpd_df[gpd_df['Country Name'].isin(sea_countries)]

# Create line plot
plt.figure(figsize=(12, 6))

# Plot line for each country with different colors
for country in sea_countries:
    country_data = sea_gdp[sea_gdp['Country Name'] == country]
    plt.plot(country_data['Year'], country_data['Value'], label=country, linewidth=2)

plt.title('GDP Over Time - Southeast Asian Countries', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('GDP', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.style.use('dark_background')
plt.show()
```

```python
# libraries
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# initiate figure
fig, ax = plt.subplots(figsize=(10, 8), dpi=120)

def update(frame):
    ax.clear()
    ax.scatter(
      1+frame, 10+frame*10,
      s=600, alpha=0.5,
      edgecolors="black"
    )
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 100)
    return fig, ax

ani = FuncAnimation(fig, update, frames=range(10))
ani.save("my_animation.gif", fps=5)
```

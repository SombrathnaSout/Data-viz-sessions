```python
days=[1,2,3,4,5,6,7] # List of days representing the time period (Day of the week)
sales_1 =[160, 150, 140,145, 175, 160, 180] # Sales data for Company 1
sales_2 = [70, 90,160, 150, 140,145, 175] # Sales data for Company 2

import matplotlib.pyplot as plt # Import the matplotlib library for plotting

plt.figure(figure= (6,5), dpi = 100) # Create a new figure with specified size and resolution
plt.plot(days, sales_1, linewidth = 1, markersize = 10, marker = 'o', linestyle = '--', label = 'Company 1') # Plot sales data for Company 1 with specific style
plt.plot(days, sales_2, marker = '*', markersize = 10, color = 'red', linestyle = '-', label = 'Company 2') # Plot sales data for Company 2 with specific style
plt.grid(color = 'green') # Add a grid to the plot with green lines
plt.title("Sales Over time", fontsize= 14, color = 'red') # Set the title of the plot with specific font size and color
plt.xlabel('Day of the week', fontsize= 15, color = 'blue') # Label the x-axis with specific font size and color
plt.ylabel('Sales', fontsize = 15, color = 'purple') # Label the y-axis with specific font size and color
# plt.fill_between(days, sales_1, color = 'blue', alpha = 0.3) # Fill area under the sales_1 line with blue color
# plt.axhline(150, color = 'black', linewidth = 3) #  Draw a horizontal line at y=150
plt.ylim(50,190) # Set the limits for the y-axis
# plt.yticks([50, 70, 90 , 110,130,150, 170,190]) #  Set specific y-axis tick marks
plt.legend(loc = 'lower right') # Add a legend to the plot at the lower right position
plt.show() # Display the plot
```

```python
# Import required library
import matplotlib.pyplot as plt

# Data: GDP values (in billions USD) from 2018-2022
years = [2018, 2019, 2020, 2021, 2022]
gdp_indonesia = [1042, 1119, 1058, 1186, 1319]    # IDN
gdp_thailand = [506, 544, 501, 505, 495]          # THA
gdp_singapore = [375, 372, 345, 397, 466]         # SGP
gdp_philippines = [346, 376, 361, 394, 404]       # PHL
gdp_vietnam = [245, 261, 271, 363, 409]           # VNM

# Create figure with specific size
plt.figure(figsize=(10, 6), dpi=100)

# Plot lines for each country with different colors and styles
plt.plot(years, gdp_indonesia, marker='o', linewidth=2, label='Indonesia (IDN)', color='#FF0000')
plt.plot(years, gdp_thailand, marker='s', linewidth=2, label='Thailand (THA)', color='#00FF00')
plt.plot(years, gdp_singapore, marker='^', linewidth=2, label='Singapore (SGP)', color='#0000FF')
plt.plot(years, gdp_philippines, marker='D', linewidth=2, label='Philippines (PHL)', color='#FFA500')
plt.plot(years, gdp_vietnam, marker='v', linewidth=2, label='Vietnam (VNM)', color='#800080')

# Customize the plot
plt.title('GDP of Southeast Asian Countries (2018-2022)', fontsize=14, pad=15)
plt.xlabel('Year', fontsize=12)
plt.ylabel('GDP (Billions USD)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Display the plot
plt.show()
```

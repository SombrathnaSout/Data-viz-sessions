```python
import seaborn as sns
import pandas as pd

# Load the Iris dataset from Seaborn's built-in dataset library
iris_data = sns.load_dataset('iris')

# Preview the data
print(iris_data.head())
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(15, 5), dpi = 60)

axes[0].hist(iris_data['petal_length'], bins=15, color='skyblue', edgecolor='black')
axes[0].set_title('Petal Length Distribution', fontsize=12)
axes[0].set_xlabel('Petal Length (cm)', fontsize=10)
axes[0].set_ylabel('Frequency', fontsize=10)
axes[0].grid(axis='y', linestyle='--', alpha=0.7)
```

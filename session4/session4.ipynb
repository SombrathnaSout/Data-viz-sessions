### How many plots you arrange in a figure 

```python
# Import matplotlib.pyplot for plotting functionality
import matplotlib.pyplot as plt
# Import numpy for numerical operations
import numpy as np

# Create x-coordinates for first plot using numpy array
x = np.array([0,1,2,3])
# Create y-coordinates for first plot using numpy array 
y = np.array([3,5,8,1])

# Create subplot in position 1 of a 1x2 grid (1 row, 2 columns, position 1)
plt.subplot(1,2,1)
# Plot x vs y coordinates for first subplot
plt.plot(x,y)

# Create x-coordinates for second plot
x = np.array([0,1,2,3])
# Create y-coordinates for second plot with larger values
y = np.array([10,20,30,40])

# Create subplot in position 2 of the 1x2 grid (1 row, 2 columns, position 2)
plt.subplot(1,2,2)
# Plot x vs y coordinates for second subplot
plt.plot(x,y)

# Display both subplots in the figure
plt.show()
```

```python
# Import matplotlib.pyplot for creating plots (already imported above)
# Import numpy for numerical operations (already imported above)

# Create x-coordinates as numpy array for first plot
x = np.array([0,1,2,3])
# Create y-coordinates as numpy array for first plot
y = np.array([3,5,8,1])

# Create subplot in position 1 of a 2x1 grid (2 rows, 1 column, position 1)
plt.subplot(2,1,1)
# Plot with blue circles ('b') and dashed lines ('--') connecting them ('o')
plt.plot(x,y,'bo--')

# Create x-coordinates as numpy array for second plot
x = np.array([0,1,2,3])
# Create y-coordinates as numpy array for second plot with larger values
y = np.array([10,20,30,40])

# Create subplot in position 2 of the 2x1 grid (2 rows, 1 column, position 2)
plt.subplot(2,1,2)
# Plot with red stars ('r*') and solid lines ('-') connecting them
plt.plot(x,y,'r*-')

# Display both subplots in the figure
plt.show()
```

```python
# Create x-coordinates array with 4 values from 0 to 3
x = np.array([0, 1, 2, 3])
# Create y-coordinates array with 4 corresponding values
y = np.array([3, 8, 1, 10])
# Create first subplot in 2x3 grid (2 rows, 3 cols, position 1)
plt.subplot(2, 3, 1)
# Plot x vs y coordinates as line graph
plt.plot(x,y)
# Add title "Sales" with large font size of 18
plt.title("Sales", fontsize = 18)

# Create new x-coordinates array for second plot
x = np.array([0, 1, 2, 3])
# Create y-coordinates with larger values for second plot
y = np.array([10, 20, 30, 40])

# Create second subplot in 2x3 grid (position 2)
plt.subplot(2, 3, 2)
# Plot x vs y coordinates
plt.plot(x,y)
# Add title for second subplot
plt.title("Product performance")

# Create x-coordinates for third plot
x = np.array([0, 1, 2, 3])
# Create y-coordinates for third plot
y = np.array([3, 8, 1, 10])

# Create third subplot in 2x3 grid (position 3)
plt.subplot(2, 3, 3)
# Plot x vs y coordinates
plt.plot(x,y)

# Create x-coordinates for fourth plot
x = np.array([0, 1, 2, 3])
# Create y-coordinates for fourth plot
y = np.array([10, 20, 30, 40])

# Create fourth subplot in 2x3 grid (position 4)
plt.subplot(2, 3, 4)
# Plot x vs y coordinates
plt.plot(x,y)

# Create x-coordinates for fifth plot
x = np.array([0, 1, 2, 3])
# Create y-coordinates for fifth plot
y = np.array([3, 8, 1, 10])

# Create fifth subplot in 2x3 grid (position 5)
plt.subplot(2, 3, 5)
# Plot x vs y coordinates
plt.plot(x,y)

# Create x-coordinates for sixth plot
x = np.array([0, 1, 2, 3])
# Create y-coordinates for sixth plot
y = np.array([10, 20, 30, 40])

# Create sixth subplot in 2x3 grid (position 6)
plt.subplot(2, 3, 6)
# Plot x vs y coordinates
plt.plot(x,y)

# Add main title for entire figure with custom color (#34ebc9) and large font size
plt.suptitle("My Yearly Sales Report", color = '#34ebc9', fontsize = 20)
# Adjust spacing between subplots - h_pad controls vertical space, w_pad controls horizontal space
plt.tight_layout(h_pad= 1 , w_pad= 3) # specify high and width between each subplot
# Display the complete figure with all subplots
plt.show()
```

# Scatterplot

```python
# Create x-coordinates array with random data points for scatter plot
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
# Create y-coordinates array with corresponding y-values for each x point
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

# Create scatter plot using x,y coordinates and set point color to red
plt.scatter(x,y, color = 'red')  # scatter() creates dot plot instead of connected lines
# Display the plot on screen
plt.show()  # Required to render the plot in notebook
```

```python
# Dataset 1: Create arrays for first product category
x1 = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])  # x-coordinates for first dataset (product quantities)
y1 = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])  # y-coordinates for first dataset (product prices)

# Create scatter plot for first dataset with green dots
plt.scatter(x1, y1, color='green')  # scatter() creates dots instead of lines, color parameter sets dot color

# Dataset 2: Create arrays for second product category 
x2 = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])  # x-coordinates for second dataset (product quantities)
y2 = np.array([90,76,97,8,121,46,23,87,94,18,97,35,86])  # y-coordinates for second dataset (product prices)

# Create scatter plot for second dataset with red dots
plt.scatter(x2, y2, color='red')  # Add second dataset to same plot with different color
plt.grid()  # Add grid lines to plot for better readability
plt.title("product category comparison")  # Add title to describe plot contents
plt.show()  # Display the final plot with both datasets
```

## Subplot for other types of graph other than line plot

```python
# Create figure with 2 vertically stacked subplots (2 rows, 1 column) that share x and y axes
fig, axes = plt.subplots(2,1, figsize=(14, 6), sharey=True, sharex=True )  # figsize=(width,height)

# Create x-coordinates array for first dataset (product quantities)
x1 = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
# Create y-coordinates array for first dataset (product prices) 
y1 = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

# Create x-coordinates array for second dataset (product quantities)
x2 = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
# Create y-coordinates array for second dataset (product prices)
y2 = np.array([90,76,97,8,121,46,23,87,94,18,97,35,86])

# Create scatter plot in first subplot (top) with blue dots
axes[0].scatter(x1,y1, color='blue')  # axes[0] refers to first/top subplot

# Create scatter plot in second subplot (bottom) with orange dots  
axes[1].scatter(x2,y2, color='orange')  # axes[1] refers to second/bottom subplot

# Adjust spacing between subplots automatically
plt.tight_layout()  # Prevents subplot titles and labels from overlapping
# Display the complete figure with both subplots
plt.show()  # Required to render the plot in notebook
```

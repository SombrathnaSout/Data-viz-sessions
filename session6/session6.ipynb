```python
import matplotlib.pyplot as plt
import seaborn as sns

iris_data = sns.load_dataset('iris')
iris_data.head()
```

# Combining multiple together and extract insight from it

```python
mean_petal_width = iris_data.groupby('species')['petal_width'].mean()
mean_petal_width.info()
```

```python
fig, ax = plt.subplots(2,2, figsize=(12,10), dpi = 90)

# Histogram of Sepal Length
ax[0,0].hist(iris_data['sepal_length'], bins=10, color=sns.color_palette('tab10')[4], edgecolor = 'black')
ax[0,0].set_title('Distribution of sepal length of all flower species')
ax[0,0].set_xlabel('Sepal lenght in cm')
ax[0,0].set_ylabel('how often it happens in the dataset')

# Relationship between Sepal length vs Sepal Width by Scatter plot

ax[0,1].scatter(iris_data['sepal_length'],iris_data['sepal_width'] , color=sns.color_palette('tab10')[3], edgecolor = 'black')
ax[0,1].set_title('Relationship between Sepal length vs Sepal Width by Scatter plot')
ax[0,1].set_xlabel('Sepal lenght in cm')
ax[0,1].set_ylabel('Sepal width in cm')

# Bar chart of Mean Petal Width of each species


ax[1,0].bar(mean_petal_width.index , mean_petal_width.values , color=['lightblue', 'green', 'red'], edgecolor = 'black')
ax[1,0].set_title('Average petal width of each species')
ax[1,0].set_xlabel('Species')
ax[1,0].set_ylabel('average petal width')

# enhancement of plot 1
ax[1,1].hist(iris_data['sepal_length'], bins=15, color=sns.color_palette('tab10')[6], edgecolor = 'black', alpha = 0.5)
ax[1,1].set_title('Distribution of sepal length of all flower species')
ax[1,1].set_xlabel('Sepal lenght in cm')
ax[1,1].set_ylabel('how often it happens in the dataset')
ax[1,1].grid(axis = 'y')

plt.tight_layout(h_pad= 2.2, w_pad= 2.3)
```

```python
fig, ax = plt.subplots(figsize=(12,5))

sns.scatterplot(x='sepal_length', y = 'sepal_width', data = iris_data, hue='species', palette='tab10')
fig.suptitle('Sepal_length vs Sepal_width by species')
plt.show()

```

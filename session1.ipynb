```python
import matplotlib.pyplot as plt
```

```python
sales_1 = [12.4, 38.84, 87, 12, 34.8, 12.5, 98.3] # Sales by week day

day_of_week = [1,2,3,4,5,6,7]

plt.figure(figsize=(10,10), dpi = 100)

plt.plot(day_of_week, sales_1, color = 'blue', marker = '*', markersize = 12)
plt.title("Sales over time")
plt.xlabel("Day")
plt.ylabel("Sales in $")
plt.grid()
plt.show()
```

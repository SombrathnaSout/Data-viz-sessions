```python
years = [2010, 2011, 2012, 2013, 2014]
values = [10,15,20,25,30]

import matplotlib.pyplot as plt
plt.plot(years, values, color = 'red')
plt.title("Simple Line Plot")
plt.xlabel("Years")
plt.ylabel("Values")
plt.grid()
plt.show()
```

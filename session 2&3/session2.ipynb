```python
# Import pandas library for data manipulation and analysis
# pd is the standard alias used for pandas
import pandas as pd  # pd: alias for pandas, used for DataFrame operations, data analysis, and manipulation
```

```python
# Read the CSV file 'movies_data.csv' into a pandas DataFrame named 'data'
# pd.read_csv(): reads CSV files, returns DataFrame
# Parameter: "movies_data.csv" is the input file path
data = pd.read_csv("movies_data.csv")

# Display the first row of the DataFrame 'data'
# head(): shows first n rows, default n=5
# Parameter: 1 means show only first row
data.head(1)
```

```python
# Display comprehensive information about the DataFrame 'data'
# info(): shows index, data types, memory usage, and non-null values for each column
# No parameters needed - displays all columns by default
data.info()
```

```python
# Check for missing values in each column of DataFrame 'data'
# isnull(): returns boolean DataFrame indicating which values are missing (True = missing)
# sum(): counts number of missing values in each column
# Returns: Series showing count of null values per column
data.isnull().sum()
```

```python
# Remove all rows with missing values (NaN) from DataFrame 'data'
# dropna(): drops rows containing any null values
# inplace=True: modifies original DataFrame instead of creating copy
data.dropna(inplace=True)
```

```python
# Check for missing values in DataFrame 'data' and sum them by column
# isnull(): returns True for missing values, False otherwise
# sum(): adds up True values (missing) for each column
# Returns: Series showing count of null values per column
data.isnull().sum()
```

```python
# Display last 5 rows of DataFrame 'data'
# tail(): shows last n rows of DataFrame
# Parameter: 5 specifies to show last 5 rows
# Returns: DataFrame with last 5 rows
data.tail(5)
```

```python
# Access the 'Released_Year' column from DataFrame 'data'
# data[column_name]: extracts single column as Series
# Parameter: "Released_Year" specifies which column to select
data["Released_Year"]
```

```python
# Convert 'Released_Year' column to numeric values
# pd.to_numeric(): converts strings/mixed types to numbers
# errors='coerce': converts invalid parsing to NaN instead of raising error
data['Released_Year'] = pd.to_numeric(data['Released_Year'], errors='coerce')
```

```python
# Convert 'Released_Year' column to datetime format
# pd.to_datetime(): converts values to datetime objects
# Parameter: data['Released_Year'] - column to convert
# Parameter: format='%Y' - specifies input is 4-digit year format (e.g. 1994)
# Returns: datetime Series with dates set to Jan 1st of each year
data['Released_Year'] = pd.to_datetime(data['Released_Year'], format='%Y')
```

```python
# Import pandas library for data manipulation and analysis
import pandas as pd
# Import matplotlib.pyplot for creating visualizations and plots 
import matplotlib.pyplot as plt

# Read CSV file 'movies_data.csv' into pandas DataFrame named 'movies'
# Parameter: 'movies_data.csv' - name of input CSV file
movies = pd.read_csv('movies_data.csv')
# Create subset of first 10 rows (index 0-9) and select only 'Series_Title' and 'IMDB_Rating' columns
# Parameter: loc[0:9] - selects rows 0 through 9 inclusive
# Parameter: ['Series_Title', 'IMDB_Rating'] - list of column names to include
subset = movies.loc[0:9, ['Series_Title', 'IMDB_Rating']]
```

```python
# Create subset of first 10 movies with Series_Title, IMDB_Rating and Runtime columns
# Parameter loc[0:9]: Select rows 0-9, Parameter ['Series_Title',...]: Select specific columns
subset = movies.loc[0:9, ['Series_Title', 'IMDB_Rating', 'Runtime']]

# Create figure with width=12 inches and height=6 inches
# Parameter figsize=(12,6): Set figure dimensions in inches
plt.figure(figsize=(12, 6))

# Plot IMDb ratings line with blue circular markers
# Parameters: x=movie titles, y=ratings, label=legend text, color=blue, marker='o' for circles
plt.plot(subset['Series_Title'], subset['IMDB_Rating'], label='IMDb Rating', color='blue', marker='o')

# Plot runtime line with orange x markers on same graph
# Parameters: x=movie titles, y=runtime, label=legend text, color=orange, marker='x' for x marks
plt.plot(subset['Series_Title'], subset['Runtime'], label='Runtime (minutes)', color='orange', marker='x')

# Set plot title with larger font size
# Parameter fontsize=16: Set title text size to 16pt
plt.title('Comparison of IMDb Ratings and Runtime', fontsize=16)

# Label x-axis as "Movie Title" with medium font size
# Parameter fontsize=12: Set label text size to 12pt
plt.xlabel('Movie Title', fontsize=12)

# Label y-axis as "Values" with medium font size
# Parameter fontsize=12: Set label text size to 12pt
plt.ylabel('Values', fontsize=12)

# Rotate x-axis labels 45 degrees and align to right edge
# Parameters: rotation=45 for angle, ha='right' for horizontal alignment
plt.xticks(rotation=45, ha='right')

# Add light gray dotted gridlines
# Parameters: color=gray, linestyle=':' for dotted, linewidth=0.5 for thin lines
plt.grid(color='gray', linestyle=':', linewidth=0.5)

# Add legend with medium font size
# Parameter fontsize=12: Set legend text size to 12pt
plt.legend(fontsize=12)

# Adjust spacing to prevent label cutoff
# No parameters - automatically adjusts subplot parameters for tight fit
plt.tight_layout()

# Display the plot
# No parameters - shows the current figure
plt.show()
```

```python
# Select first 10 rows and specific columns from movies dataframe
# Parameters: loc[0:9] selects rows 0-9, column names select those specific columns
subset = movies.loc[0:9, ['Series_Title', 'IMDB_Rating', 'Runtime']]

# Create figure and primary axis with specified size
# Parameters: figsize=(width,height) in inches
fig, ax1 = plt.subplots(figsize=(12, 6))

# Set color for first plot
# Parameter: color1 stores the color string 'blue' for reuse
color1 = 'blue'

# Set x-axis label with formatting
# Parameters: fontsize=12 sets text size to 12pt
ax1.set_xlabel('Movie Title', fontsize=12)

# Set primary y-axis label with color matching the data
# Parameters: color matches line color, fontsize=12 for text size
ax1.set_ylabel('IMDb Rating', color=color1, fontsize=12)

# Plot IMDb ratings on primary y-axis
# Parameters: x=movie titles, y=ratings, color=blue, marker='o' for circles, label for legend
line1 = ax1.plot(subset['Series_Title'], subset['IMDB_Rating'], 
                 color=color1, marker='o', label='IMDb Rating')

# Color the y-axis ticks to match the data
# Parameters: axis='y' specifies y-axis, labelcolor sets tick color
ax1.tick_params(axis='y', labelcolor=color1)

# Create secondary y-axis sharing x-axis with primary
# Parameter: twinx() creates new axis sharing x-axis
ax2 = ax1.twinx()

# Set color for second plot
# Parameter: color2 stores the color string 'orange' for reuse
color2 = 'orange'

# Set secondary y-axis label with color matching data
# Parameters: color matches line color, fontsize=12 for text size
ax2.set_ylabel('Runtime (minutes)', color=color2, fontsize=12)

# Plot runtime on secondary y-axis
# Parameters: x=movie titles, y=runtime, color=orange, marker='x' for x marks, label for legend
line2 = ax2.plot(subset['Series_Title'], subset['Runtime'], 
                 color=color2, marker='x', label='Runtime')

# Color the secondary y-axis ticks to match data
# Parameters: axis='y' specifies y-axis, labelcolor sets tick color
ax2.tick_params(axis='y', labelcolor=color2)

# Combine plot lines for unified legend
# Parameter: + operator concatenates the line lists
lines = line1 + line2

# Extract labels from each line for legend
# Parameter: get_label() retrieves label text from each line
labels = [l.get_label() for l in lines]

# Create legend combining both plots
# Parameters: lines=line objects, labels=text, loc=position, fontsize=text size
ax1.legend(lines, labels, loc='upper right', fontsize=12)

# Set overall plot title
# Parameter: fontsize=16 sets title text size
plt.title('Comparison of IMDb Ratings and Runtime', fontsize=16)

# Set x-axis tick labels with rotation
# Parameters: rotation=45 sets angle, subset provides labels
ax1.set_xticklabels(subset['Series_Title'], rotation=45)

# Rotate and align x-axis tick labels
# Parameters: rotation=45 sets angle, ha='right' aligns text
plt.xticks(rotation=45, ha='right')

# Add grid lines to plot
# Parameters: color=gray, linestyle=':' for dotted, linewidth=0.5 for thin lines
plt.grid(color='gray', linestyle=':', linewidth=0.5)

# Adjust plot layout to prevent label cutoff
# Parameter: tight_layout() automatically adjusts spacing
plt.tight_layout()

# Display the plot
# Parameter: show() displays the current figure
plt.show()
```

```python
# Import required libraries for plotting and data manipulation
import matplotlib.pyplot as plt  # For creating visualizations
import pandas as pd  # For data analysis and manipulation

# Read CSV file into pandas DataFrame
# Parameter: file path string
movies_data = pd.read_csv("movies_data.csv")

# Convert Released_Year to datetime format and handle errors
# Parameters: errors='coerce' converts invalid values to NaT, format='%Y' specifies year format
movies_data['Released_Year'] = pd.to_datetime(movies_data['Released_Year'], errors='coerce', format='%Y')

# Remove rows with missing Released_Year values
# Parameter: subset specifies columns to check for NaN
movies_data = movies_data.dropna(subset=['Released_Year'])

# Split genre strings into lists at comma+space
# Parameter: str.split(', ') splits string on comma followed by space
movies_data['Genre'] = movies_data['Genre'].str.split(', ')

# Create separate row for each genre in a movie
# Parameter: column name to explode
exploded_data = movies_data.explode('Genre')

# Group by year and count genres
# Parameter: size() counts occurrences in each group
genre_by_year = exploded_data.groupby('Released_Year').size()

# Create figure with specified dimensions
# Parameter: figsize=(width, height) in inches
plt.figure(figsize=(12, 6))

# Create line plot with markers
# Parameters: x=years, y=counts, marker='o' for dots, linestyle='-' for solid line, color='b' for blue
plt.plot(genre_by_year.index, genre_by_year.values, marker='o', linestyle='-', color='b')

# Set plot title with font size
# Parameters: title text string, fontsize=14 for text size
plt.title('Total Number of Genres Released Per Year', fontsize=14)

# Label x-axis with font size
# Parameters: label text, fontsize=12 for text size
plt.xlabel('Year', fontsize=12)

# Label y-axis with font size
# Parameters: label text, fontsize=12 for text size
plt.ylabel('Total Genres', fontsize=12)

# Add grid lines to plot
# Parameters: True enables grid, linestyle='--' for dashed lines, alpha=0.6 for opacity
plt.grid(True, linestyle='--', alpha=0.6)

# Adjust spacing between plot elements
# Parameter: automatically adjusts padding between plot elements
plt.tight_layout()

# Display the plot
# Parameter: shows the current figure
plt.show()
```

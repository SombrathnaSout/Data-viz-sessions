```python
from scipy.interpolate import make_interp_spline, interp1d, UnivariateSpline, PchipInterpolator

years = pivot_df.index
values = pivot_df.values.T

custom_colors= ["#003F5C","#304C7E","#665190","#9F5095","#D44E86","#F95D6A","#FF7B42","#FFA600"]

pivot_smooth_interp_spline = pd.DataFrame({col : make_interp_spline(years, pivot_df[col])(years_new) for col in pivot_df.columns}, index=years_new)

pivot_smooth_linear = pd.DataFrame({col : interp1d(years, pivot_df[col], kind='linear')(years_new) for col in pivot_df.columns}, index=years_new)

pivot_smooth_cubic = pd.DataFrame({col : interp1d(years, pivot_df[col], kind='cubic')(years_new) for col in pivot_df.columns}, index=years_new)

pivot_smooth_univariate_spline = pd.DataFrame({col : UnivariateSpline(years, pivot_df[col])(years_new) for col in pivot_df.columns}, index=years_new)

pivot_smooth_pchip = pd.DataFrame({col : PchipInterpolator(years, pivot_df[col])(years_new) for col in pivot_df.columns}, index=years_new)

pivot_smooth_linear.info()

```

```python
def plot_stackplot(ax,x,y, labels, title):
    ax.stackplot(x, y, labels=labels, colors=custom_colors)
    ax.legend(loc='upper left')
    ax.set_title(title)

fig, ax = plt.subplots(5, 1, figsize=(10, 30), dpi = 90)

plot_stackplot(ax[0], pivot_smooth_interp_spline.index, pivot_smooth_interp_spline.values.T , pivot_df.columns ,'Total wealth by country (interp spline)')
plot_stackplot(ax[1], pivot_smooth_linear.index, pivot_smooth_linear.values.T , pivot_df.columns,'Total wealth by country (linear)')
plot_stackplot(ax[2], pivot_smooth_cubic.index, pivot_smooth_cubic.values.T , pivot_df.columns,'Total wealth by country (cubic)')
plot_stackplot(ax[3], pivot_smooth_univariate_spline.index, pivot_smooth_univariate_spline.values.T ,pivot_df.columns, 'Total wealth by country (univariate spline)')
plot_stackplot(ax[4], pivot_smooth_pchip.index, pivot_smooth_pchip.values.T , pivot_df.columns, 'Total wealth by country (pchip)')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6), dpi = 90)
plt.stackplot(pivot_smooth_interp_spline.index, pivot_smooth_interp_spline.values.T, labels=pivot_df.columns, colors=custom_colors)
ax = plt.gca()
years = [2000,2005,2010,2015,2021]

y_values = [pivot_df.loc[year].sum() for year in years]

plt.vlines(years, 0, [y*1.05 for y in y_values], colors='black', linewidth = 1)
plt.plot(years, [y*1.05 for y in y_values], 'o', color='black')

for year, y_value in zip(years, y_values):
    x_offset = -3 if year == 2021 else -1.5
    y_offset = 20000 if year == 2021 else 26000
    plt.text(year + x_offset, y_value + y_offset, f'${y_value:,.0f}', fontsize=10, fontweight='bold')


ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.tick_params(left=False, labelleft=False)

plt.text(2000, 320000,
         'Aggregated \nHousehold \nWealth', # Title ('\n' allows you to go to the line)
         fontsize=40, # High font size for style
         color='black',
         fontweight = 'bold')
plt.show()
```
